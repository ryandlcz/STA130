### 5. Attributes vs Methods / Function Calls

```python
titanic_df.describe          #attribute (no parentheses)
```

```python
titanic_df.describe()        #method/function call (with parentheses)
```

- Attribute  
An attribute simply holds or returns a value, it does not perform any action. Thus, attributes are characteristics or properties of an object. In the context of a pandas data frame, df.describe simply refrences the describe method, but doesn't execute it. As such, the info df.describe() contains when refrencing our data set is displayed as raw, unformatted text.

- Methods  
Methods perform an action when called upon, which can be done by adding parentheses() after a function name which executes the function. When df.describe() is ran, the python function is executed and the information from df.describe() when refrencing our data set is formatted and displayed neatly in a table.

### 6. Definitions of rows provided by describe() method

```python
import pandas as pd

# Load the dataset from the URL
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
titanic_df = pd.read_csv(url)

titanic_df.describe()
```

##### Definitions + examples of df.describe() rows according to ChatGPT:
- count: number of non-null (non-NaN) entries in the column
    - ex. if a column has 100 rows but 2 rows contain missing values (NaN), *count* = 98.
    
- mean: average of the values in the column, calculated as the sum of all values divided by the *count*.
    - ex. if the values in a column are [2, 3, 5], the mean would be (2+3+5)/3 = 3.33
    
- std (standard deviation): measure of the dipersion/spread of the values from the mean. it indicates how much the values typically deviate from the mean.
    - ex. low std: values are closely packed around the mean while
    - ex. high std: values are spread out further from mean
    
- min: smallest (minimum) value in the column
    - ex. if values in a column are[2, 5, 3], the *min* would be 2.
    
- 25% (1st quartile / Q1): refers to the 25th percentile of data, the value below which 25% of the data fall.
    - ex. if 25% of the values in a column <=2, the 25% value would be 2.

- 50% (2nd quartile / Q2): refers to the 50th percentile of data, half of the values are below this point, while half are above. also known as the **median**.
    - ex. if values in a column are [1, 3, 5] the median would be 3.
    
- 75% (3rd quartile / Q3): refers to the 75th percentile of the data, the value below which 75% of the dat fall.
    - ex. if 75% of the values in a column >=5, the 75% value would be 5.
    
- max: largest (maximum) value in the column
    - ex. if values in a column are [2, 3, 5], the *max* would be 5.


#### What does df.describe() do with missing values?

- When df.describe() runs in pandas, it automatically **ignores missing values (NaNs)** when calculating summary statistics.

- Count reflects only number of non-NaN values in each column
- All other summary stats are calculated based on available values (NaN's are ignored)


### 7. Using df.dropna() or del df['col'] to maximize efficiency

```python
titanic_df.dropna()
```

- df.dropna()
    - Removes rows or columns cointaing missing values, allowing one to retain as much valid data as poppsible in the Data Frame.
    - Can selectively remove only rows where a specific column (or set of columns) has missing values using options like *subset* or *how*.


```python
#df.dropna()  # Drops rows with any missing values
titanic_df.dropna(subset=['pclass', 'age'])  # Drops rows only if 'col1' or 'col2' has missing values
```

```python
titanic_df.shape    #confirming titanic_df still contains full original data
```

##### df.dropna() over del df['col']
- As shown above when running attribute df.shape, the titanic data frame has a total of 891 rows. However, when df.dropna() is ran, only 182 rows are displayed as rows as all rows with missing data is dropped.
- An advantage of df.dropna(), is that we can **selectively remove rows** as shown above with (subset=['pclass', 'age'] as it drops rows only if 'col1' (pclass) and 'col2' (age) have missing values. Thus, 714 rows are now displayed when subset is ran.

##### del df['col'] over df.dropna()

- One might choose del df['col'] over df.dropna() to remove an entire column regardless of NaN's.
- Entire columns can be removed to **maximize efficiency** and **simplify data preprocessing** if not needed (e.g. identifier, metadata, any other irrelevant feature) or if a column contains significant number of missing values to the point where available data is sparse not usable.
- dropna() may remove too many rows in the case of a column having many missing values. In this case, simply removing a problematic column with **del** may be better to **preserve rows of valuable data** in other columns.

##### Importance of applying del df['col'] before using df.dropna()

- Primary factor is  of **reduction of data loss**
    - Which helps maintain data integrity & simplifying data cleaning.
    - If a particular column has high number of missing values, applying dropna() first may cause an unnecessary drop in rows resulting in signicant loss of useful data from other columns
    - Using del df['col'] first ensures avoiding dropping rows that might contain valuable info in other columns.
    - ex. If 'col' has ~80% missing values, using df.dropna() *first* would remove ~80% of dataset. easily avoided by simply deleting the problematic column before dropping rows.
